# DFT: Frequenzgang eines Moving Average Filters (Musterlösung)


```python
import os, sys
module_path = os.path.abspath(os.path.join('..')) # append directory one level up to import path
if module_path not in sys.path:  # ... if it hasn't been appended already
    sys.path.append(module_path)
    
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

figsize = {"figsize": (12,5), "facecolor":"w"}
```

Setzt man $h_{MA,3} = \{1;1;1;1\}$ periodisch fort, erhält man eine Gleichspannung, so wie sie das Ergebnis der DFT beschreibt. Die drei Nullstellen sind auch die Nullstellen des Frequenzgangs des MA-Filters, die Betrachtungen wiedersprechen sich daher nicht.

Um den Betrag der komplexen Werte der DFT darzustellen, berechnet man dere Betrag mit `np.abs()`.

Die Skalierung ist korrekt wenn man ein endliches Signal annimmt (Impulsantwort), interpretiert man das Eingangssignal als Ausschnitt aus einem unendlich ausgedehnten Signal (hier: DC), muss man durch die Anzahl der DFT-Punkte dividieren.

Für ein aufschlussreicheres Ergebnis wendet man zero-padding an, d.h. man berechnet die DFT mit mehr Punkten des Signals mit einer entsprechenden Zahl angehängter Nullen. Der Plot zeigt den Frequenzgang zwischen $0 \ldots f_S$ an.

```python
h = np.ones(4) # Geben Sie hier die Länge des MA-Filters (Ordnung + 1) an.
n = np.arange(len(h))
fig = plt.figure(1, **figsize)
fig.clf()
ax1 = fig.add_subplot(211)
ax1.stem(n,h)
ax1.set_xlabel(r"$n \; \rightarrow$")
ax1.set_ylabel(r"$h[n] \; \rightarrow$")

ax2 = fig.add_subplot(212)
H = np.abs(np.fft.fft(h,256))
k = np.arange(len(H))
ax2.stem(k,H)
ax2.set_xlabel(r"$k \; \rightarrow$")
ax2.set_ylabel(r"$H[k] \; \rightarrow$")

fig.tight_layout()
plt.show()
```
